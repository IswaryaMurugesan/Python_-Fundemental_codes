**Data Structures - Strings & Tuples**

**1.String Concatenation**

```python
Name =input("Enter the name :")

string1 = "Hello "
string2 = Name

## Concatenate string1 & string2
Text1 = string1 +string2

string3 = " Welcome to Python Programming"

Text2 = Text1 + string3
print(Text2)

## Concate string 3

print(f"\nconcatenate String :",string1,string2,string3 )
```

**2. String Slicing and Indexing:**

```python

##a. Print the first character of the string.

print("First Character:",Text2[0])

## b.Print the last character of the string.

print("Last character :",Text2[-1] )

## c. Print the first 5 characters of the string.
print("First 5 Characters :",Text2[0:5])

## d. Print the last 11 characters of the string.
print("Last 11 characters :", Text2[-11:])

## e. Print the string in reverse.
print("Reverse string :",Text2[::-1])
```

**3. String Methods:**

```python
strM = " Python beginner tutorial "

print("\n----String Methods-----")

print("\n strM = Python beginner tutorial")

## a. Convert the sentence to uppercase.

print("Upper Case :",strM.upper())

## b. Convert the sentence to lowercase.

print("Lower case :",strM.lower())

## c. Use Capitalize and return the sentence to the original input form.

print("Capitalized :", strM.capitalize())

## d. Count the total number of occurrences of character ‘t’ in the string.

print("Count of total number of character 't' :",strM.count("t"))

## e. Replace all occurrences of “Python” with “Data Analytics” in the input string

print("Replace String :",strM.replace("Python","Data Analytics"))
```

**Tuples (Creation, Modification and Access) :**

```python

Tuple1 = (10,20,30)
Tuple2 =(40,50,60)

## a. Concatenate the two tuples and store it in “t_combine”

t_combine =Tuple1+Tuple2

print("\n**** Tuple Operations ****")
print("\nConcatenate:",t_combine)

## b. Repeat the elements of “t_combine” 3 times

print("Repeated tuples : ", t_combine *3)

## c. Access the 3rd element from “t_combine”

print("Access the 3rd element :",t_combine[2])

## d. Access the first three elements from “t_combine”

print("Access the first 3 elements :",t_combine[0:3])

## e. Access the last three elements from “t_combine”

print("Access the last 3 element :",t_combine[-3:])
```
